```python
import matplotlib.pyplot as plt
```

```python
f = open('times.txt', 'r')
```

```python
lines = []
for line in f:
    lines.append(line)
```

```python
lines
```

```python
file_start = {}
file_end = {}
file_accuracy = {}
file_lines = {}
file_deltas = {}
```

```python
for line in lines:
    month,day,time,fname = line.split()[5:]
    timestr = "2017-02-{} {}".format(day,time)
    file_end[fname] = timestr
    print(fname,timestr)
```

```python
for fname in file_end:
    with open('experiments/' + fname, 'r') as ff:
        f_lines = []
        for line in ff:
            f_lines.append(line)
        file_lines[fname] = f_lines
```

```python
with open('foo.txt', 'a+') as a:
    a.write('YOOO\n')
a.close()
```

```python
for fname,lines in file_lines.items():
    file_start[fname] = lines[1]
    acc = (lines[-1][10:]).split()
    n,p = acc[0],acc[-1]
    file_accuracy[fname] = (float(n) / float(p))
#     for line in lines[:2]:
#         print(line)
```

```python
print(file_start)
print(file_end)
print(file_accuracy)
```

```python
from datetime import datetime
```

```python
for fn in file_start:
    start,end = file_start[fn],file_end[fn]
#     print(start[:16],end)
    start = datetime.strptime(start[:16],'%Y-%m-%d %H:%M')
    end = datetime.strptime(end, '%Y-%m-%d %H:%M')
    file_deltas[fn] = (end-start)
```

```python
def out(fname):
    ret = "#"*40 + '\n\n'
    ret += "end accuracy      : {}".format(file_accuracy[fname])
    ret += "\ncomputation time  : {}".format(file_deltas[fname])
    return ret + "\n"
```

```python
ev = open('everything.csv', 'w')
```

```python
for fname in file_end:
    with open('experiments/' + fname, 'a+') as ff:
        ff.write(out(fname))
    ff.close()
```

```python
everything = {}
for fname in file_end:
    everything[fname] = {
        'start' : file_start[fname],
        'end' : file_end[fname],
        'delta' : file_deltas[fname],
        'accuracy' : file_accuracy[fname]
    }
```

```python
headers = ['filename', 'start', 'end', 'delta', 'accuracy']
print(",".join(headers),file=ev)
```

```python
for k,v in everything.items():
    line = [k] + [vv for kk,vv in v.items()]
    print(",".join([str(e).replace('\n','') for e in line]),file=ev)
```

```python
ev.close()
```

```python
file_epoch_acc = {}
```

```python
for fname in file_end:
    epoch_acc = []
    for line in file_lines[fname][11:]:
        x = (line.split(':')[1]).split()
        xx = [float(e) for e in x[::2]]
        epoch_acc.append(xx[0] / xx[1])
    file_epoch_acc[fname] = epoch_acc
```

```python
typ = ['-','--', '-.']
idx=0
for fn,ea in file_epoch_acc.items():
    lab = fn[4:-4]
    m,r,l = lab.split('_')
    m,r,l = m[1:],r[1:],l[1:]
    plt.plot(ea, typ[int(idx/6)],
             label="mb: {:3}, lr: {:5}, layers: {:13}".format(m,r,l),
            linewidth=1.50)
    idx+=1

plt.xlabel('Epochs')
plt.ylabel('Classification Rate')
plt.title('Accuracy Over Epochs')
# plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.legend(loc=4)
plt.rcParams['figure.figsize'] = [14.0,14.0]
# plt.style.use('seaborn-dark-palette')
plt.style.use('fivethirtyeight')
# plt.style.use('bmh')
plt.rcParams['font.family'] = 'monospace'
plt.gca().set_xlim([0,60])
plt.gca().set_ylim([0,1])
plt.show()
```
